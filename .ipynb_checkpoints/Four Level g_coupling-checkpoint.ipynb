```python
import matplotlib.pyplot as plt
import numpy as np
import qutip as qt
import scipy
```

# Test 2: Four Level Capacitively Coupled iSwap

##### Relevant Unit Conversions
$10^9hz=1ghz, 1s=10^9 ns, 1s=10^6\mu s$

```python
π = np.pi
µs_to_ns = 1e3
```

### States
$|g\rangle=(1,0,0,0)^T$, $|e\rangle=(0,1,0,0)^T$, $|f\rangle=(0,0,1,0)^T$, $|n\rangle=(0,0,0,1)^T$

```python
num_states = 4

g_state = qt.basis(num_states, 0)
e_state = qt.basis(num_states, 1)
f_state = qt.basis(num_states, 2)
n_state = qt.basis(num_states, 3)
```

### Operators
$a,a^\dagger, b, b^\dagger$ 

```python
a = qt.tensor(qt.destroy(num_states), qt.qeye(num_states)) #lower first qubit
adag = a.dag() #raise first qubit
b = qt.tensor(qt.qeye(num_states), qt.destroy(num_states)) #loser second qubit
bdag = b.dag() #raise second qubit

gg_proj = qt.tensor(g_state, g_state).proj()
ge_proj = qt.tensor(g_state, e_state).proj()
gf_proj = qt.tensor(g_state, f_state).proj()
gn_proj = qt.tensor(g_state, n_state).proj()

eg_proj = qt.tensor(e_state, g_state).proj()
ee_proj = qt.tensor(e_state, e_state).proj()
ef_proj = qt.tensor(e_state, f_state).proj()
en_proj = qt.tensor(e_state, n_state).proj()

fg_proj = qt.tensor(f_state, g_state).proj()
fe_proj = qt.tensor(f_state, e_state).proj()
ff_proj = qt.tensor(f_state, f_state).proj()
fn_proj = qt.tensor(f_state, n_state).proj()

ng_proj = qt.tensor(n_state, g_state).proj()
ne_proj = qt.tensor(n_state, e_state).proj()
nf_proj = qt.tensor(n_state, f_state).proj()
nn_proj = qt.tensor(n_state, n_state).proj()
```

### Hamiltonian
Three Level: $$H/\hbar = \sum_{i=1}^2{\left(\omega_i b_i^\dagger b_i + \frac{\alpha_i}{2} b_i^\dagger b_i^\dagger b_i b_i\right)} + g (b_1^\dagger b_2+b_1 b_2^\dagger).$$

Four Level: $$H/\hbar = \sum_{i=1}^2{\left(\omega_i b_i^\dagger b_i + \frac{\alpha_i}{2} b_i^\dagger b_i^\dagger b_i b_i + \frac{\alpha_i^2}{4} b_i^\dagger b_i^\dagger b_i^\dagger b_i b_i b_i\right)} + g (b_1^\dagger b_2+b_1 b_2^\dagger).$$

```python
def hamiltonian(omega_1, omega_2, alpha_1, alpha_2, g):
    t1 = (omega_1 * adag * a) + (omega_2 * bdag * b)
    t2 = (alpha_1 / 2 * adag * adag * a * a) + (alpha_2 / 2 * bdag * bdag * b * b)
    t3 = (alpha_1 * alpha_1 / 4 * adag * adag * adag * a * a * a)
    t4 = (alpha_2 * alpha_2 / 4 * bdag * bdag * bdag * b * b * b)
    t5 = g * (adag * b + bdag * a)
    return t1 + t2 + t3 + t4 + t5
```

### Plotting Function

```python
def plotresult(time_evolution_b):
    ge_pops = qt.expect(ge_proj, time_evolution_b.states)
    eg_pops = qt.expect(eg_proj, time_evolution_b.states)

    gf_pops = qt.expect(gf_proj, time_evolution_b.states)
    fg_pops = qt.expect(fg_proj, time_evolution_b.states)

    gn_pops = qt.expect(gn_proj, time_evolution_b.states)
    ng_pops = qt.expect(ng_proj, time_evolution_b.states)
    
    fe_pops = qt.expect(fe_proj, time_evolution_b.states)
    ef_pops = qt.expect(ef_proj, time_evolution_b.states)
    
    en_pops = qt.expect(en_proj, time_evolution_b.states)
    ne_pops = qt.expect(ne_proj, time_evolution_b.states)
    
    fn_pops = qt.expect(fn_proj, time_evolution_b.states)
    nf_pops = qt.expect(nf_proj, time_evolution_b.states)

    gg_pops = qt.expect(gg_proj, time_evolution_b.states)
    ee_pops = qt.expect(ee_proj, time_evolution_b.states)
    ff_pops = qt.expect(ff_proj, time_evolution_b.states)
    nn_pops = qt.expect(nn_proj, time_evolution_b.states)
    
    fig, ax = plt.subplots(tight_layout=True)
    ax.plot(t_vec * µs_to_ns, ge_pops, label="ge")
    ax.plot(t_vec * µs_to_ns, eg_pops, label="eg")

    ax.plot(t_vec * µs_to_ns, gf_pops, label="gf")
    #ax.plot(t_vec * µs_to_ns, fg_pops, label="fg")

    #ax.plot(t_vec * µs_to_ns, gn_pops, label="gn")
    #ax.plot(t_vec * µs_to_ns, ng_pops, label="ng")

    ax.plot(t_vec * µs_to_ns, fe_pops, label="fe")
    #ax.plot(t_vec * µs_to_ns, ef_pops, label="ef")
    
    #ax.plot(t_vec * µs_to_ns, en_pops, label="en")
    #ax.plot(t_vec * µs_to_ns, ne_pops, label="ne")

    #ax.plot(t_vec * µs_to_ns, fn_pops, label="fn")
    #ax.plot(t_vec * µs_to_ns, nf_pops, label="nf")

    ax.plot(t_vec * µs_to_ns, gg_pops, label="gg")
    ax.plot(t_vec * µs_to_ns, ee_pops, label="ee")
    #ax.plot(t_vec * µs_to_ns, ff_pops, label="ff")
    #ax.plot(t_vec * µs_to_ns, nn_pops, label="nn")

    ax.legend()
    ax.set_xlabel('Time, $t$ [ns]')
    ax.set_ylabel(r'Population, $P_{mn}$')
    ax.set_title('SWAP gate')

    plt.show()
    plt.close()
```

## On Resonance, Same Anharmonicity

```python
omega = 2*π * 6000
g = 2*π * 8
alpha = 2*π * -300

h_ww_aa = hamiltonian(omega_1=omega, omega_2=omega, alpha_1=alpha, alpha_2=alpha, g=g)

t_start = 0
t_end = 3 * π / (2 * g)
t_vec = np.linspace(t_start, t_end, 601)

psi0 = qt.tensor(g_state, e_state)

time_evolution_b = qt.mesolve(h_ww_aa, psi0, t_vec, [], [])

plotresult(time_evolution_b)
```

## Slightly Detuned, Same Anharmonicity

```python
omega1 = 2*π * 6000 #6 Ghz converted to microsecond timescale
detuning = 5 # Mhz/2π
omega2 = omega1 - (2*π* detuning)
g = 2*π * 8
alpha = 2*π * -300

h_ww_aa = hamiltonian(omega_1=omega1, omega_2=omega2, alpha_1=alpha, alpha_2=alpha, g=g)

t_start = 0
t_end = 3 * π / (2 * g)
t_vec = np.linspace(t_start, t_end, 601)

psi0 = qt.tensor(g_state, e_state)

time_evolution_b = qt.mesolve(h_ww_aa, psi0, t_vec, [], [])

plotresult(time_evolution_b)
```
